### Series from lists

```python
# string

country = ['India' ,'Japan' ,'France' ,'RSA','USA']

pd.Series(country)
```

```python
# integer

runs = [134,105,65,84,50]
pd.Series(runs)
```

```python
# series has two things
# 1.value and 2.index(it is by deault genrated by pandas, 0 based)
# dtype = data type of series value(object = 99 % case me string hi hoga)
```

```python
# custom index (khud ka index)

# you can override the deafult pandas index

marks = [40,57,89,100]
subject = ['DL' , 'SE' , 'Robotics' , 'Ml']

pd.Series(marks,index = subject)
```

```python
# series name

marks = pd.Series(marks,index = subject,name = 'Nitish sir ke marks')
print(marks)
```

### Series from dict

```python
# you can make series using the dictionary

marks = {
    'Maths' : 100,
    'English' : 89,
    'Science' : 70,
    'History' : 100
}

marks_series = pd.Series(marks,name = 'nitish sir ke marks')
marks_series

# by using dict you will the get same result that you get using index parameter while you created series using list.
```
