###`Q-6` Write a code which can display the bar chart of top 5 teams who have played maximum number of matches in the IPL.
>  Hint: Be careful the data is divided in 2 different cols(Team 1 and Team 2)


````python
# code here
````

````python
one = (df['Team1'].value_counts()).sort_index().values
````

````python
ones = (df['Team1'].value_counts()).sort_index().index
````

````python
Team_name = ones
````

````python
two = df['Team2'].value_counts().sort_index().values
````

````python
Total_matches_played = one + two
````

````python
Total_matches_played
````

````python
Total_matches = pd.DataFrame()
````

````python
Total_matches['Total_matches_played'] = Total_matches_played
````

````python
Total_matches
````

````python
Total_matches['Team_name'] = Team_name
````

````python
Total_matches
````

````python
Total_matches.set_index('Team_name',inplace=True)
````

````python
new_df = Total_matches['Total_matches_played'].sort_values(ascending=False).head()
````

````python
new_df.plot(kind = 'bar')
````

###`Q-8:` Team1 vs Team2 Dashbord
Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.
```
team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
```

````python
# code here
def WinLossRecordAndPOM(team1 , team2):
  print( df[ (df['Team1'] == team1) & (df['Team2'] == team2) | (df['Team1'] == team2) & (df['Team2'] == team1)]['TossWinner'].value_counts(),'\n')
  print( df[ (df['Team1'] == team1) & (df['Team2'] == team2) | (df['Team1'] == team2) & (df['Team2'] == team1)]['Player_of_Match'].value_counts().head())



````
