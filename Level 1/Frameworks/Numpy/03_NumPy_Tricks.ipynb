#### np.sort()

```python
# returns a sorted numpy array
```

```python
import numpy as np
```

```python
a = np.random.randint(1,100,15)
a
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
np.sort(a) # ascending
```

```python
np.sort(a)[::-1] # descending
```

```python
np.sort(b) # by default row wise
```

```python
np.sort(b,axis = 0) # column wise
```
