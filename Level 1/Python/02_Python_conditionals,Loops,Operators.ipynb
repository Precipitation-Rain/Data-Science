# Operators

```python
# 1] Arithemetic Operators

#used for mathematical operation
print(4 + 5)
print(4 - 5)
print(4 * 5)
print(4 / 5)
print(4 // 5) #integer division -> converts result of division into integer
print(42 % 5)
print(2**3) # power operator
```

```python
# 2] Relational Operators

#also called as comparision operators
print(4 > 5)
print(4 < 5)
print(4 >= 5)
print(4 <= 5)
print(4 == 4)
print(4 != 4)
```

```python
# 3] Logical Operators

# used for logic building and condition checking
print (1 and 0)

print(1 or 0)

print(not 1)
```

```python
# 4] Bitwise Operatoes

# -> works on binary numbers

#1)Bitwise and
print(2 & 3)

#2)Bitwise or
print(2 | 3)

#3)Bitwise xor
print(2 ^ 3)

#4)Bitwise not
print(~3)

#5)Bitwise leftshift
print(4 >> 2)

#6)Bitwise rightshift
print(4 << 2)
```

```python
# 5] Assignment Operator

a = 2 # = is assignment operator -> used to assign values
print(a)

#Assignment operator uses with many arithematic operator

a += 2
print(a)

a -= 2
print(a)

a *= 2
print(a)

a /= 2
print(a)

a %= 2
print(a)
```

```python
# 6] Membership Operator

# in / not in  -> works on string,list,tuple,dictionary,set.

print('D' in 'Delhi')

print(2 not in [3,4,5])
```

```python
# python has two types of loop
# 1.while loop
# 2.for loop
```

```python
#while loop
#print table of user inputed number

number = int(input("Enter the value of number "))

i=1;

while (i<=10):
  print(number,"*",i,"=",number*i)
  i=i+1
```

```python
# while loop with else statement

i=1

while(i<3):
  print(i)
  i+=1
else:
  print("loop ended")
```

```python
#Guessing game
```

```python
import random
jackpot = random.randint(1,100)

guess = int(input("Guess the number "))
counter = 1

while (guess != jackpot):
  if guess < jackpot :
    print("Galat ! guess higher")

  else :
    print("Galat ! guess lower")

  guess = int(input("guess again "))
  counter += 1

else:
  print("You guessed right.","It is ",jackpot)
```
